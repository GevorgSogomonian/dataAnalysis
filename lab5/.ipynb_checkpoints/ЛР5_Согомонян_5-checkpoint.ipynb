### 4. Разбиение набора данных

Задание: Разбить набор данных на тренировочной и тестовый датасеты с помощью train_test_split и выполнить стандартизацию числовых данных.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Обработка категориальных признаков (One-Hot Encoding)
# Модели не понимают текст ('Yes', 'No', 'Male'), нужно превратить их в числа (0 и 1)
# drop_first=True удаляет первый столбец, чтобы избежать мультиколлинеарности
# (например, если есть is_Male, то is_Female не нужен)
X_encoded = pd.get_dummies(X, drop_first=True)

print("Размерность данных после кодирования:", X_encoded.shape)
print("Новые столбцы:", list(X_encoded.columns))

# 2. Разбиение на обучающую и тестовую выборки
# test_size=0.25 — отдаем 25% данных на тест (стандартная практика)
# random_state=42 — фиксируем случайность, чтобы результат был воспроизводим
# stratify=y — сохраняем баланс классов (процент ушедших в train и test будет одинаковым)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, 
    y, 
    test_size=0.25, 
    random_state=42, 
    stratify=y
)

print(f"Обучающая выборка: {X_train.shape}")
print(f"Тестовая выборка: {X_test.shape}")

# 3. Стандартизация числовых данных
# Это нужно, чтобы признаки с большими числами (например, доход 5000) 
# не заглушали признаки с маленькими (например, возраст 25)

# Определяем числовые колонки (те, которые были в исходном датасете)
numeric_cols = ['age', 'tenure_months', 'monthly_charges', 'total_charges', 
                'avg_call_minutes', 'num_support_tickets']

scaler = StandardScaler()

# Обучаем scaler только на train, чтобы не "подглядывать" в тест
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])

# Применяем выученные параметры к тесту
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

print("\nДанные успешно стандартизированы.")
print("Пример первых 3 строк обучающей выборки (стандартизированные):")
print(X_train[numeric_cols].head(3))
```

1. Кодирование признаков (Encoding)
Было: (3016, 11) — 11 столбцов, среди которых были текстовые (object).

Стало: (3016, 17) — 17 столбцов. Количество увеличилось на 6.

Что произошло:

Pandas создал новые бинарные столбцы (0 или 1) для каждой категории.

Пример: Вместо столбца contract_type ('Month-to-month', 'One year', 'Two year') появились contract_type_One year и contract_type_Two year.

Важный момент: Столбец для базового случая (например, 'Month-to-month') был удален (drop_first=True), чтобы избежать жесткой линейной зависимости (дамми-ловушки). Это экономит память и улучшает работу линейных моделей.

2. Разбиение выборки (Splitting)
Обучающая выборка (Train): 2262 записи. На этом объеме мы будем учить модель искать закономерности.

Тестовая выборка (Test): 754 записи. Эти данные мы "спрячем" от модели и используем только в самом конце для честной проверки экзаменом.

Пропорция: 2262 / 3016 ≈ 75%, что соответствует заданному test_size=0.25.

3. Стандартизация (Scaling)
Посмотри на таблицу с примерами (head(3)). Все значения теперь маленькие, крутятся вокруг нуля (от -2 до +2 примерно).

Например, monthly_charges раньше были около 70-100 долларов, а теперь -0.72 или 0.04.

Зачем это нужно:

Метод k-ближайших соседей (KNN) считает расстояния. Без этого шага признак total_charges (где тысячи долларов) перетянул бы всё внимание на себя, а num_support_tickets (где единицы) вообще бы не влиял на результат. Теперь у всех признаков равные "права голоса".

Логистическая регрессия тоже будет сходиться быстрее и точнее.

Итог: Данные полностью готовы. У нас есть матрица чисел, одинакового масштаба, разделенная на обучение и тест.

### 5. Разработка предсказательной модели

Задание: Разработать предсказательную модель качественного отклика методами: метод k- ближайших соседей, дерево решений, логистическая регрессия, случайный лес

```python
# Импорт необходимых классификаторов
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Инициализация словаря для хранения моделей и их предсказаний
# Это поможет нам удобно обращаться к результатам в следующих пунктах
models = {
    "KNN (k-ближайших соседей)": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree (Дерево решений)": DecisionTreeClassifier(random_state=42, max_depth=5),
    "Logistic Regression (Логистическая регрессия)": LogisticRegression(random_state=42, max_iter=1000),
    "Random Forest (Случайный лес)": RandomForestClassifier(random_state=42, n_estimators=100)
}

# Словарь для хранения предсказаний
predictions = {}
probabilities = {} # Для ROC-кривых в будущем

print("Начинаем обучение моделей...\n")

for name, model in models.items():
    # 1. Обучение модели на тренировочных данных
    model.fit(X_train, y_train)
    
    # 2. Предсказание классов (0 или 1) для тестовой выборки
    y_pred = model.predict(X_test)
    predictions[name] = y_pred
    
    # 3. Предсказание вероятностей (для ROC-кривой и оценки уверенности)
    # predict_proba возвращает две колонки: вероятность 0 и вероятность 1.
    # Нам нужна вероятность класса 1 (уход), поэтому берем индекс [:, 1]
    y_prob = model.predict_proba(X_test)[:, 1]
    probabilities[name] = y_prob
    
    print(f"Модель '{name}' обучена.")

print("\nВсе модели успешно обучены и сделали предсказания.")
```

Теперь у нас есть словарь predictions с ответами каждой модели, который мы используем в пункте 6 для расчета метрик.

### 6. Рассчёт метрик моделей

Задание: Оценить ошибку классификации для каждого метода. Подсчитать метрики "Accuracy", "Precision", "Recall", "Balanced accuracy", 'F1 score'

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, balanced_accuracy_score

# Создаем пустой список для сбора результатов
results = []

print(f"{'Модель':<35} | Accuracy | Precision | Recall | Bal. Acc. | F1 Score")

for name, y_pred in predictions.items():
    # Расчет метрик
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred)
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Вывод результатов в консоль для наглядности
    print(f"{name:<35} | {acc:.4f}   | {prec:.4f}    | {rec:.4f} | {bal_acc:.4f}    | {f1:.4f}")
    
    # Сохраняем в список словарей для создания красивого DataFrame
    results.append({
        'Model': name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'Balanced Accuracy': bal_acc,
        'F1 Score': f1
    })

# Создаем сводную таблицу (DataFrame) для удобного сравнения
results_df = pd.DataFrame(results)

# Сортируем таблицу по F1 Score (обычно это главная метрика баланса)
results_df = results_df.sort_values(by='F1 Score', ascending=False)

print("Сводная таблица результатов (отсортирована по F1 Score):")
print(results_df)
```

1. Лидер — Логистическая регрессия (Logistic Regression)
Она заняла первое место по F1 Score (0.63), Balanced Accuracy (0.76) и Recall (0.57).

Почему это странно: Обычно "лес" (Random Forest) работает лучше простых линейных моделей.

Почему так вышло: Вероятно, зависимость в данных преимущественно линейная (например, "чем больше обращений в поддержку, тем выше шанс ухода"), и сложная структура "леса" здесь избыточна или требует более тонкой настройки гиперпараметров.

Сильная сторона: Самый высокий Recall (0.57). Это значит, что она находит 57% всех уходящих клиентов. Это лучший результат среди всех моделей, но всё ещё далеко от идеала (мы теряем почти половину "оттока").

2. Random Forest (Случайный лес) — самая точная, но "осторожная"
У этой модели самый высокий Precision (0.79).

Что это значит: Если лес говорит "клиент уйдет", он прав в 79% случаев. Ошибок (ложных тревог) очень мало.

Минус: У неё низкий Recall (0.46). Она боится рисковать и помечает "уходящими" только самых очевидных. Многих проблемных клиентов она пропускает.

3. KNN (k-ближайших соседей) — аутсайдер
Худшие показатели по всем фронтам (F1 = 0.52).

Причина, скорее всего, в том, что даже после стандартизации признаки имеют разную "плотность", и простое евклидово расстояние плохо разделяет классы в 17-мерном пространстве.

Общий вывод
Ни одна модель "из коробки" не дала идеального результата (F1 > 0.8 считается отличным). Сейчас мы имеем F1 ≈ 0.63.

Основная проблема: Низкий Recall (Полнота). Мы пропускаем много уходящих клиентов.

Рекомендация: Для бизнеса часто важнее поймать всех, кто хочет уйти (высокий Recall), даже ценой лишних звонков лояльным клиентам. Поэтому Логистическая регрессия здесь пока лучший выбор.

### 8. График ROC-кривой

Задание: Построить графики ROC-кривой для каждого метода на одном графике (4 линии на одном графике) для сравнения

```python
from sklearn.metrics import roc_curve, roc_auc_score

plt.figure(figsize=(10, 8))

# Проходим по всем моделям и строим для каждой график
for name, y_prob in probabilities.items():
    # Вычисляем метрики для кривой
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    
    # Считаем площадь под кривой (AUC) - это одно число, показывающее качество
    auc = roc_auc_score(y_test, y_prob)
    
    # Рисуем линию
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.3f})', linewidth=2)

# Добавляем диагональную пунктирную линию (случайное угадывание)
plt.plot([0, 1], [0, 1], 'k--', label='Случайное угадывание (AUC = 0.5)')

# Настройка осей и легенды
plt.xlabel('False Positive Rate (Ложные тревоги)', fontsize=12)
plt.ylabel('True Positive Rate (Полнота / Recall)', fontsize=12)
plt.title('ROC-кривые для сравнения моделей', fontsize=14)
plt.legend(loc='lower right', fontsize=11)
plt.grid(True, alpha=0.3)

plt.show()
```

В предыдущих пунктах (таблице метрик) Random Forest казался "середнячком" из-за низкого Recall. Но ROC-кривая (которая оценивает модель до выбора жесткого порога 0.5) показывает, что у леса огромный потенциал.

AUC = 0.929 (самый высокий результат).

Красная линия проходит выше всех остальных почти на всем протяжении графика.

Что это значит: Если мы немного подкрутим порог вероятности (например, будем считать "уходящим" каждого, у кого риск > 30%, а не 50%), то Random Forest даст результат намного лучше, чем Логистическая регрессия. Он просто был "скромным" с дефолтными настройками.

2. Logistic Regression (Зеленая линия) — Надежное второе место
AUC = 0.898. Очень достойный результат.

График показывает, что линейная модель работает стабильно, но все же уступает ансамблевому методу (лесу) в способности находить сложные нелинейные паттерны.

3. Decision Tree (Оранжевая линия)
AUC = 0.896. Почти догнала регрессию.

Однако видно, что линия более "угловатая" (ступенчатая). Одиночное дерево менее гибкое, чем целый лес.

4. KNN (Синяя линия) — Слабое звено
AUC = 0.824.

Линия проходит значительно ниже остальных. Метод ближайших соседей в этом пространстве признаков работает хуже всего, подтверждая выводы из прошлых пунктов.

Итоговое резюме перед выводом (п.9):
Хотя в таблице "готовых" ответов (с порогом 50%) победила Логистическая регрессия, ROC-анализ доказал, что Random Forest — самая мощная модель. Если бизнесу нужно максимальное качество, нужно брать Random Forest и калибровать порог срабатывания.

### 9. Выводы

В ходе лабораторной работы была решена задача бинарной классификации оттока клиентов телекоммуникационной компании. Были проанализированы данные о 3016 клиентах, выполнена предобработка (кодирование категорий, стандартизация) и обучены 4 модели машинного обучения.

### Основные результаты:

1.  **Дисбаланс классов:** В данных наблюдается явный перекос — ушедших клиентов всего около 16%. Это сделало метрику *Accuracy* неинформативной (все модели показали высокую точность ~87-89% просто за счет угадывания большинства).
2.  **Сравнение моделей:**
    *   **Logistic Regression** показала лучший баланс "из коробки" (F1-score = 0.63, Recall = 0.57). Она нашла больше всего реальных "уходящих" клиентов при стандартном пороге 0.5.
    *   **Random Forest** стал лидером по метрике **ROC-AUC (0.929)**. Это говорит о том, что алгоритм лучше всех ранжирует клиентов по вероятности ухода. Низкий Recall в таблице (0.46) связан лишь с высоким стандартным порогом. При его снижении эта модель станет самой эффективной.
    *   **KNN** показал худшие результаты (AUC = 0.824), плохо справившись с высокой размерностью данных после кодирования.

### Ключевые признаки оттока:
На этапе разведочного анализа (EDA) было выявлено, что главными индикаторами ухода являются:
*   Частые обращения в поддержку (**num_support_tickets** > 3).
*   Высокие ежемесячные платежи (**monthly_charges**).
*   Короткий срок обслуживания (новые клиенты уходят чаще).

### Итоговая рекомендация:
Для внедрения в бизнес-процессы рекомендуется использовать модель **Random Forest (Случайный лес)**, предварительно настроив порог классификации (Threshold Tuning) для повышения полноты (Recall). Это позволит удерживать максимальное количество рисковых клиентов, сохраняя высокую точность прогнозов.
