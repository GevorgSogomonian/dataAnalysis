### 1.Чтение файла и предварительная обработка данных

#### 1.1 Чтение файла

```python
# импорт библиотек, чтение файла с помощью pandas
import pandas as pd
df = pd.read_csv('credit_risk2.csv')
df.info()
df.describe()
```

Итак, из вывода df.info() мы видим, что есть много нулевых значений для Rate и Emp_length. Из вывода df.describe() мы видим выбросы в столбцах Age и Emp_length.

#### 1.2 Удаление пропусков

Пропуски в Emp_length мы удалим, в Rate заменим на среднее значение, так как там пропущено ~10% от всех данных, что очень много для простого удаления.

```python
# Удаляем пропуски в Emp_length
df = df.dropna(subset=['Emp_length'])
# Заполняем пропуски в Rate средними значениями
df.loc[:, 'Rate'] = df['Rate'].fillna(df['Rate'].mean())
df.info()
df.describe()
```

#### 1.3 Проверка дубликатов

Удаление явных дубликатов

```python
# Ищем явные дубликаты
df.duplicated()
```

```python
# Удаляем явные дубликаты
df = df.drop_duplicates().reset_index(drop=True)
# Проверяем результат
df.duplicated()
```

Удаление неявных дубликатов

```python
# Проверка столбца Home
df['Home'].value_counts()
```

```python
# Удаление неявных дулбикатов столбца Home
df.loc[:, 'Home'] = df['Home'].replace('RENET', 'RENT')
df['Home'].value_counts()
```

```python
# Проверка столбца Intent
df['Intent'].value_counts()
```

```python
# Проверка столбца Default
df['Default'].value_counts()
```

```python
# Удаление неявных дулбикатов столбца Default
df.loc[:, 'Default'] = df['Default'].replace('No', 'N')
df['Default'].value_counts()
```

#### 1.4 Проверка типов данных

```python
df.info()
```

Типы данных корректны

### 2. Построение и анализ точечной диаграммы

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(nrows=2, ncols=3, figsize = (18, 12))

# Точечная диаграмма Emp_length(Age) by Status
sns.scatterplot(data=df, x='Age', y='Emp_length', hue='Status', ax=ax[0,0], alpha=0.6)
ax[0,0].set_title("Emp_length(Age) by Status")
ax[0,0].set_xlim(20, 30)
ax[0,0].set_ylim(0, 20)

# Точечная диаграмма Amount(Income) by Cred_length
sns.scatterplot(data=df, x='Income', y='Amount', hue='Cred_length', ax=ax[0,1], alpha=0.6)
ax[0,1].set_title("Amount(Income) by Cred_length")
ax[0,1].set_xlim(0,100000)
ax[0,1].set_ylim(0,25000)

# Точечная диаграмма Percent_income(Rate) by Default
sns.scatterplot(data=df, x='Rate', y='Percent_income', hue='Default', ax=ax[1,0], alpha=0.6)
ax[1,0].set_title("Percent_income(Rate) by Default")

# Точечная диаграмма Income(Amount) by Status
sns.scatterplot(data=df, x='Amount', y='Income', hue='Status', ax=ax[1,1], alpha=0.6)
ax[1,1].set_title("Income(Amount) by Status")

# Точечная диаграмма Income(Rate) by Intent
sns.scatterplot(data=df, x='Rate', y='Income', hue='Intent', ax=ax[0,2], alpha=0.6)
ax[0,2].set_title("Income(Rate) by Intent")

# Точечная диаграмма Income(Percent_income) by Home
sns.scatterplot(data=df, x='Percent_income', y='Income', hue='Home', ax=ax[1,2], alpha=0.6)
ax[1,2].set_title("Income(Percent_income) by Home")
```

Анализ графиков

- Emp_length(Age) by Status
    Прослеживается прямая зависимость максимального стажа работы от возраста. Закономерности по Status нет.
  
- Amount(Income) by Cred_length
    Есть зависимость размера кредита от дохода: заёмы больше ~20000 берут по большей части люди с доходом >40000,
    люди с доходом <20000 в основном берут кредиты до ~7000. Закономерности по Cred_length нет.

- Income(Amount) by Intent
    Закономерностей нет.

- Percent_income(Rate) by Default
    Люди, которые раньше допускали дефолт по кредиту в основном имеют ставку >12%, а кто не допускал <12%.

- Income(Amount) by Status
    Люди которые подают заявку на кредит >20000 с доходом <100000 обычно имеют статус одобрения '1'(что бы это не значило),
    а с доходом >100000 '0'

- Income(Percent_income) by Home
    Для заёмщиков с домом в аренду прослеживается обратная зависимость Income от Percent_income. Большинство заёмщиков с Percent_income <0.2 имеют доход >150000

### 3. Построение гистограммы

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15,15))

ax[0,0].hist(df['Age'], bins=50, width=0.8)
ax[0,0].set_title("Age")

ax[0,1].hist(df['Income'], bins=10, width=20000)
ax[0,1].set_title("Income")

ax[0,2].hist(df['Home'], bins=4, width=0.2)
ax[0,2].set_title("Home")

ax[1,0].hist(df['Emp_length'], bins=100, width=1)
ax[1,0].set_title("Emp_length")

ax[1,1].hist(df['Intent'], bins=6, width=0.5)
ax[1,1].set_title("Intent")

ax[1,2].hist(df['Amount'], bins=10, width=2000)
ax[1,2].set_title("Amount")

ax[2,0].hist(df['Rate'], bins=10, width=0.5)
ax[2,0].set_title("Rate")

ax[2,1].hist(df['Status'], bins=10, width=0.5)
ax[2,1].set_title("Status")

plt.show()
```
